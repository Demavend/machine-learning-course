2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f2 = sp.sin(x)
f2_prime = sp.diff(f2, x)
print(f"Derivative of sin(x): {f2_prime}")
```

```python
f3 = sp.ln(x)
f3_prime = sp.diff(f3, x)
print(f"Derivative of ln(x): {f3_prime}")
```

```python
f2_func = sp.lambdify(x, f2, modules=['numpy'])
f2_prime_func = sp.lambdify(x, f2_prime, modules=['numpy'])

f3_func = sp.lambdify(x, f3, modules=['numpy'])
f3_prime_func = sp.lambdify(x, f3_prime, modules=['numpy'])

x2_vals = np.linspace(-2*np.pi, 2*np.pi, 400)
x3_vals = np.linspace(0.1, 10, 400)

y2 = f2_func(x2_vals)
y2_prime = f2_prime_func(x2_vals)

y3 = f3_func(x3_vals)
y3_prime = f3_prime_func(x3_vals)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x2_vals, y2, label='sin(x)')
plt.plot(x2_vals, y2_prime, label="sin'(x) = cos(x)", linestyle='--')
plt.title('Function and Derivative of sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x3_vals, y3, label='ln(x)')
plt.plot(x3_vals, y3_prime, label="ln'(x) = 1/x", linestyle='--')
plt.title('Function and Derivative of ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### Summary

The derivatives were calculated using SymPy:

- For `f(x) = sin(x)`, the derivative is `f'(x) = cos(x)`.
- For `f(x) = ln(x)`, the derivative is `f'(x) = 1/x`.

The plots show that:
- `sin(x)` oscillates between -1 and 1, while its derivative `cos(x)` also oscillates but is phase-shifted by π/2.
- `ln(x)` increases slowly for large values of `x`, while its derivative `1/x` decreases as `x` grows, indicating that the growth rate of `ln(x)` slows down for larger `x`.


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f4 = x**2 - 5*x + 1
f4_prime = sp.diff(f4, x)
print(f"Derivative of f(x): {f4_prime}")
```

```python
f4_func = sp.lambdify(x, f4, modules=['numpy'])
f4_prime_func = sp.lambdify(x, f4_prime, modules=['numpy'])

x4_vals = np.linspace(0, 100, 200)
y4_vals = f4_func(x4_vals)
y4_prime_vals = f4_prime_func(x4_vals)

plt.plot(x4_vals, y4_vals, label='Total Cost: f(x) = x² - 5x + 1')
plt.plot(x4_vals, y4_prime_vals, label="Marginal Cost: f'(x) = 2x - 5", linestyle='--')
plt.xlabel('Production Units (x)')
plt.ylabel('Cost')
plt.title('Total Cost Function and Its Derivative')
plt.legend()
plt.grid(True)
plt.show()
```

### Summary

The cost function `f(x) = x² - 5x + 1` was analyzed. Its derivative, calculated using SymPy, is `f'(x) = 2x - 5`.

The plot shows that:
- The total cost increases quadratically as production grows.
- The derivative represents marginal cost, which increases linearly with production. For small production volumes (x < 2.5), marginal cost is negative, suggesting economies of scale. After that point, marginal cost becomes positive, indicating increasing production costs per additional unit.
